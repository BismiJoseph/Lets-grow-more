###  Number of terrorism per year

```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state',
                       'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                       'nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
x_year = df['Year'].unique()
y_count_years = df['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = x_year,
           y = y_count_years,
           palette = 'rocket')
plt.xticks(rotation = 45)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year')
plt.title('Attack_of_Years')
plt.show()
```

```python
plt.subplots(figsize=(15,6))
sns.countplot('Year',data=df,palette='RdYlGn_r',edgecolor=sns.color_palette("YlOrBr", 10))
plt.xticks(rotation=45)
plt.title('Number Of Terrorist Activities Each Year')
plt.show()
```

More number of attacks occured in the year of 2014.

```python
plt.subplots(figsize=(15,6))
sns.barplot(df['Country'].value_counts()[:15].index,df['Country'].value_counts()[:15].values,palette='Blues_d')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation= 90)
plt.show()
```

From the abive graph we came to know that Iraq is more affected by terrorist attack.

```python
sns.barplot(data=df,x=df.Year, y=df.Region)
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.countplot(x='iyear',hue='success',data=df1)

plt.xlabel('Attack Year', fontsize=20, fontweight='bold')
plt.ylabel('Number of attacks', fontsize=20, fontweight='bold')

plt.title('Trend of increase in Terror from 1970 to 2017', fontweight="bold", size=25)

plt.xticks(rotation=60, fontweight="bold", size=15)
plt.yticks(fontweight="bold", size=15)

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.barplot(x=df1['country_txt'].value_counts()[:20].index,y=df1['country_txt'].value_counts()[:20].values)

plt.xlabel('Country', fontsize=20, fontweight='bold')
plt.ylabel('Total attack counts', fontsize=20, fontweight='bold')

plt.title('Total number of attacks Country-wise', fontweight="bold", size=25)

plt.xticks(rotation=90, fontweight="bold", size=15)
plt.yticks(fontweight="bold", size=15)

plt.show()
```

From the above chart we can see that Iraq has the higgest number of terror attack and Egypt has lower

```python
plt.figure(figsize=(10,6))

sns.barplot(x=df1['targtype1_txt'].value_counts().index,y=df1['targtype1_txt'].value_counts().values)

plt.xlabel('Target type', fontsize=20, fontweight='bold')
plt.ylabel('Total attack counts', fontsize=20, fontweight='bold')

plt.title('Total count of attack on differnt target type', fontweight="bold", size=25)

plt.xticks(rotation=90, fontweight="bold", size=15)
plt.yticks(fontweight="bold", size=15)

plt.show()
```

From the above plot we can conclude that Attackers target the people who are private citizens and have property, Military person, police, Government(General) and many more
